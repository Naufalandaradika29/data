# **B. Data Preprocessing dan Transformation**

```python
# Memeriksa jika terdapat nilai null atau NaN
df.isnull().sum()[df.isnull().sum() > 0]

#Atau bisa menggunakan code ini untuk menemukan data yang bervalue 0 (bukan null)
cols_with_zeros = [col for col in df.columns if (df[col] == 0).any()]
print("Kolom dengan nilai 0:", cols_with_zeros)
```

```python
# Mengecek tipe data dari kolom yang memiliki null atau NaN
df.dtypes[df.isnull().sum() > 0]
```

```python
# Mengisi data yang null jika tipe datanya Object (tpi tidak akan kami gunakan) jika numerik menggunakan mean
# Data bertype Kategorik
df['kelas_terapi'] = df['kelas_terapi'].fillna(df['kelas_terapi'].mode()[0])
# Data bertype Numerik
df['ppn_jual'] = df['ppn_jual'].fillna(df['ppn_jual'].mean())
```

```python
# Pastikan kolom waktu bertipe datetime
df['waktu_transaksi'] = pd.to_datetime(df['waktu_transaksi'], errors='coerce')

# Mengekstrak komponen tanggal, hari, dll
df['tanggal'] = df['waktu_transaksi'].dt.date
df['hari'] = df['waktu_transaksi'].dt.day_name()
df['bulan'] = df['waktu_transaksi'].dt.month
df['is_weekend'] = df['waktu_transaksi'].dt.weekday

# Buat kolom baru untuk jam transaksi
df['jam_transaksi'] = df['waktu_transaksi'].dt.hour

# Tampilkan hasil
df[['waktu_transaksi', 'jam_transaksi']]
```

perintah di atas di gunakan untuk mengolah informasi waktu dari kolom (waktu_transaksi) seperti bulan,jam,hari,dan tanggal

```python
# Mengubah kolom qty, hj, dan ppn_jual menjadi tipe data numerik
df['qty'] = pd.to_numeric(df['qty'], errors='coerce')
df['hj'] = pd.to_numeric(df['hj'], errors='coerce')
df['ppn_jual'] = pd.to_numeric(df['ppn_jual'], errors='coerce')

# Hitung total transaksi termasuk PPN
df['total_transaksi'] = df['qty'] * df['hj'] * (1 + df['ppn_jual'])

# Opsional: isi nilai NaN dengan 0
df['total_transaksi'] = df['total_transaksi'].fillna(0)

# Tampilkan hasil
df[['qty', 'hj', 'ppn_jual', 'total_transaksi']].head()
```

```python
# Mencari data yang bertipe object
object_columns = df.select_dtypes(include='object').columns
print("Kolom bertipe object:")
print(object_columns)
```

```python
# Mencari ada berapa class dalam 1 kolom
pd.value_counts(df['no_resep'])
```

```python
# Menghitung frekuensi kemunculan setiap jam penjualan untuk mengetahui pola waktu transaksi
pd.value_counts(df['jam_jual'])
```

```python
# Mengisi missing value dengan 'unknown' dan konversi ke string
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('unknown').astype(str)
```

```python
# Mengimpor LabelEncoder untuk mengubah data kategorikal menjadi numerik
from sklearn.preprocessing import LabelEncoder

# Salin data
df_encoded = df.copy()

# Inisialisasi LabelEncoder untuk tiap kolom
label_encoders = {}
for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = df_encoded[col].astype(str)  # Konversi semua nilai ke string
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le  # Simpan encoder untuk inverse-transform

df_encoded
```

Tujuan dari kategorik ke numerik untuk memudahkan saat melakukan data training kedepannya (seingat saya -kuzzty)

# **C. Data Visualization**

```python
# Mengambil yang hanya kolom numerik
numeric_df = df.select_dtypes(include=[np.number])

# Hitung korelasi dan visualisasikan
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Korelasi antar fitur numerik")
plt.show()
```

```python
# Plot histogram pengunjung per jam
plt.figure(figsize=(12, 5))
sns.histplot(df['jam_transaksi'], bins=24, kde=False, color='skyblue')
plt.title('Frekuensi Pengunjung Berdasarkan Jam Transaksi')
plt.xlabel('Jam (0-23)')
plt.ylabel('Jumlah Transaksi')
plt.xticks(range(0, 24))
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Plot total penjualan per jam
penjualan_per_jam = df.groupby('jam_transaksi')['total_transaksi'].sum()

plt.figure(figsize=(12, 5))
penjualan_per_jam.plot(kind='bar', color='salmon')
plt.title('Total Penjualan per Jam')
plt.xlabel('Jam (0-23)')
plt.ylabel('Total Penjualan (Rp)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Hitung total jumlah terjual per nama obat
obat_terlaris = df.groupby('nama_obat')['qty'].sum().sort_values(ascending=False).head(10)

# Plot bar chart untuk 10 obat terlaris
plt.figure(figsize=(14, 6))
obat_terlaris.plot(kind='bar', color='mediumseagreen')
plt.title('10 Obat Paling Sering Terjual')
plt.xlabel('Nama Obat')
plt.ylabel('Jumlah Terjual (qty)')
plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Hitung total penjualan per obat untuk menentukan 10 teratas
total_penjualan_obat = df.groupby('nama_obat')['qty'].sum().sort_values(ascending=False)
top_10_obat = total_penjualan_obat.head(10).index

# Filter data hanya untuk 10 obat teratas
penjualan_obat_per_jam = df[df['nama_obat'].isin(top_10_obat)].groupby(['jam_transaksi', 'nama_obat'])['qty'].sum().reset_index()

# Pivot data
pivot_df = penjualan_obat_per_jam.pivot(index='jam_transaksi', columns='nama_obat', values='qty').fillna(0)

# Urutkan kolom berdasarkan total penjualan (agar plot lebih mudah dibaca)
pivot_df = pivot_df[top_10_obat]

# Plot stacked bar chart
plt.figure(figsize=(14, 6))
pivot_df.plot(kind='bar', stacked=True, colormap='tab20', figsize=(14,6))

plt.title('Histogram Penjualan 10 Obat Teratas per Jam')
plt.xlabel('Jam Transaksi (0-23)')
plt.ylabel('Jumlah Obat Terjual')
plt.xticks(rotation=0)
plt.legend(title='Nama Obat', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
