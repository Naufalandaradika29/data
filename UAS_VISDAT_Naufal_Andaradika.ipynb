```python
import matplotlib.pyplot as plt

bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des']
penjualan = [10000000, 12000000, 15000000, 14000000, 16000000, 13000000, 20000000, 22000000, 19000000, 17000000, 15000000, 14000000]

plt.figure(figsize=(10,6))
plt.plot(bulan, penjualan, marker='o', linestyle='-', color='blue')
plt.title('Tren Penjualan Bulanan Toko Elektronik')
plt.xlabel('Bulan')
plt.ylabel('Total Penjualan (Rp)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
umur = ['18-24', '25-34', '35-44', '45-54', '55+']
pria = [120, 150, 100, 80, 40]
wanita = [130, 160, 90, 60, 30]

# Menentukan posisi x
x = np.arange(len(umur))  # posisi dasar
width = 0.35  # lebar batang

# Membuat figure
plt.figure(figsize=(10,6))

# Bar chart berdampingan
plt.bar(x - width/2, pria, width, label='Pria', color='skyblue')
plt.bar(x + width/2, wanita, width, label='Wanita', color='salmon')

# Label dan tampilan
plt.xticks(x, umur)
plt.title('Distribusi Usia dan Jenis Kelamin Pengunjung Website')
plt.xlabel('Kelompok Usia')
plt.ylabel('Jumlah Pengunjung')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

merek = ['Merek A', 'Merek B', 'Merek C', 'Merek D', 'Merek E']
pangsa = [30, 25, 20, 15, 10]

plt.figure(figsize=(8,6))
bars = plt.barh(merek, pangsa, color='skyblue')
plt.title('Pangsa Pasar Smartphone di Indonesia')
plt.xlabel('Persentase (%)')

# Tambahkan label persentase di ujung batang
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{bar.get_width()}%', va='center')

plt.tight_layout()
plt.gca().invert_yaxis()  # Membalik agar yang terbesar di atas
plt.show()
```
